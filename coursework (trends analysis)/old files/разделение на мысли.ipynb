```python
filtered_db = 'C:\\Users\\busya\\Desktop\\anaconda\\проекты\\курсовая\\final.db'

conn = sqlite3.connect(filtered_db)
cursor = conn.cursor()

cursor.execute('SELECT * from vacancy')
filtered = cursor.fetchall()
conn.close()
```

```python
main_list = []
for i in filtered:
    if "strong" in i[1]:
        texttest = i[1]
        teststrong = texttest.split('<strong>')
        cur_list = [j.split('</strong>') for j in teststrong if '</strong>' in j] 
        main_list.append(cur_list)
```

```python
keywords = ['требования', 'требуется', 'если', 'ждем', 'ждём', 'ожидаем', 'ищем', 'будет', 'компетенции', 'плюсом', 'желательно'
           'знание', 'стек', 'стэк', 'технологии', 'навыки', 'обязательно', 'от вас', 'опыт', 'ожидания', 'плюсами']
```

```python
requir = []
for i in main_list:
    for j in i:
        for k in keywords:
            if k in j[0].lower():
                requir.append(j[1])
requir[1]
```

```python
work_list = []
for i in requir:
    work_list.append(i.split('>'))
```

```python
new_work_list = []
for i in work_list:
    ap_list = []
    for j in i:
        ap_list.append(j.split('<'))
    new_work_list.append(ap_list)
new_work_list[0][0][1]
```

```python
flatten_list = [] # а как какать? то есть хз как флеттен сделать
for i in new_work_list:
    for j in i:
        for k in j:
            flatten_list.append(k)
```
