```python
final = 'dataset.db'

conn = sqlite3.connect(final)
cursor = conn.cursor()

cursor.execute('SELECT * from vacancies')
final_db = cursor.fetchall()
conn.close()
```

```python
import sqlite3
import collections
import re
```

```python
dbs = [
    'sberbank_hh.db',
    'drweb_hh.db',
    'ifmo_hh.db',
    'kasper_hh.db',
    'tinkoff_hh.db'
]
```

```python
# "умная" проверка дубликатов -> теперь мы не считаем, что один id может встретиться в базе только один раз

mentions = {id_: [] for id_ in ids}

for db in dbs:
    check = 'C:\\Users\\busya\\Desktop\\anaconda\\проекты\\uni\\\курсовая\\' + db
    conn = sqlite3.connect(check)
    cursor = conn.cursor()
    cursor.execute('SELECT * from vacancy')
    check_db = cursor.fetchall()
    conn.close()
    
    for job in check_db:
        id_ = job[1].split('/')[-1]
        if id_ in mentions:
            mentions[id_].append(job[4][:10])
            
    for id_ in mentions:
        mentions[id_] = list(set(mentions[id_]))
```

```python
l = [
    y for y in [list(set(map(lambda x: x.split('-')[1], mentions[id_])))[:-1] 
    for id_ in mentions] if y != []
]
```

```python
flatten = []
for i in l:
    for j in i:
        flatten.append(j)
```

```python
l1 = list(dict(collections.Counter(flatten)).values())
```
