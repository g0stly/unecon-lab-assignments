### Проблема: выделение нужных предложений из описания вакансии
    > все объявления (даже от одного работодателя) имеют разную структуру, разделы описания профессии могут идти в разном порядке и иметь разное озаглавие, в связи с чем возникает задача - понять, как выделить нужное из всего текста.

```python
import pymorphy2
import string
import collections
import itertools
import nltk
from nltk.corpus import stopwords
import gensim
from gensim.utils import simple_preprocess
morph = pymorphy2.MorphAnalyzer()
```

```python
descrs = {}
left = 369
for job in filtered:
    id_ = job[0].split('/')[-1]
    desc = requests.get('https://api.hh.ru/vacancies/' + id_).json()['description']
    descrs[id_] = desc
    left -= 1
    print(f'осталось обработать {round(left/369 * 100, 1)} % вакансий.')
```

```python
def remove_stopwords(words):
    return [word for word in words if word not in stopwords.words('russian')]

def prepare_text(text):
    return [remove_stopwords(gensim.utils.simple_preprocess(str(sentence), deacc=True)) \
               for sentence in text]
```

```python
main_list = []

list_id = [i[0].split('/')[-1] for i in filtered]

for i in list_id:
    if "<strong>" in descrs[i]:
        texttest = descrs[i]
        teststrong = texttest.split('<strong>')
        cur_list = [i.split('</strong>') for i in teststrong if '</strong>' in i] 
        main_list.append(cur_list)
```
